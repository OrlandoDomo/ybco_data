```python
import numpy as np
import matplotlib.pyplot as plt
import matplotlib.gridspec as gridspec
import csv
import pandas as pd
from scipy.signal import find_peaks

path= 'O:\\Scripts\\Python_scripts\\data\\'

o2= ['6p','8p','10p','12p','14p','16p','20p','22p','24p','26p','30p',]
o2_horno= ['horno_10p','horno_14p','horno_16p','horno_20p','horno_30p']
```

```python
def plot_drx(df):

    plt.plot(df['Angle'], df['Intensity'])
    plt.xlabel('Angulo')
    plt.ylabel('Intensidad')
    plt.title('DRX YBCO')
    plt.xlim([10, 80])
    plt.ylim([0, 20000])
```

```python
fig, axs = plt.subplots(11, sharex= True)
fig.subplots_adjust(hspace=0.05, top= 0.96)
fig.suptitle('DRX a varios porcentajes de oxigeno')
fig.set_size_inches(20,15)

for i in range(len(o2)):
    file= path + o2[i] + '.csv'
    df= pd.read_csv(file)
    
    peaks, _ = find_peaks(df['Intensity'].values, height= 500, 
                      threshold= 10 , distance= 30, width= 3, prominence= 200)
    
    o2percent= o2[i][:-1] + '%'
    axs[i].plot(df['Angle'], df['Intensity'], color= 'black', label= o2percent)
    axs[i].vlines(df['Angle'][peaks], ymin=0, ymax=20000, ls= ':', color='r', label = 'Picos aprox')
    axs[i].set_xlim(20,80)
    axs[i].set_ylim(1000,20000)
    axs[i].legend()

for ax in axs:
    ax.label_outer()
    
```
